```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Flights dataset (time-series data)
df = sns.load_dataset("flights")

# Line Plot Function
def line_plot():
    plt.figure(figsize=(10, 5))
    sns.lineplot(data=df, x="year", y="passengers", marker="o", color="b")
    plt.title("Passenger Trends Over Years")
    plt.xlabel("Year")
    plt.ylabel("Number of Passengers")
    plt.grid()
    plt.show()

# Function call
line_plot()

```

```python


# Tips dataset (categorical data)
df_tips = sns.load_dataset("tips")

# Bar Plot Function
def bar_plot():
    plt.figure(figsize=(8, 5))
    sns.barplot(data=df_tips, x="day", y="total_bill", palette="viridis")
    plt.title("Average Total Bill per Day")
    plt.xlabel("Day")
    plt.ylabel("Total Bill")
    plt.show()

# Function call
bar_plot()
```

```python

from mpl_toolkits.mplot3d import Axes3D

# 3D Bar Plot Function
def bar_plot_3d():
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')

    x = np.arange(5)  # Dummy X values (Years)
    y = np.arange(5)  # Dummy Y values (Months)
    X, Y = np.meshgrid(x, y)
    Z = np.random.randint(100, 500, size=X.shape)  # Random Passenger Data

    ax.bar3d(X.ravel(), Y.ravel(), np.zeros_like(Z.ravel()), 1, 1, Z.ravel(), shade=True)

    ax.set_xlabel("Year")
    ax.set_ylabel("Month")
    ax.set_zlabel("Passengers")
    plt.title("3D Passenger Data")
    plt.show()

# Function call
bar_plot_3d()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = sns.load_dataset("iris")

# Sub-plots Function
def sub_plots():
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # 2x2 grid

    # Scatter Plot
    sns.scatterplot(ax=axes[0, 0], data=df, x="sepal_length", y="petal_length", hue="species")
    axes[0, 0].set_title("Scatter Plot")

    # Histogram
    axes[0, 1].hist(df["sepal_length"], bins=15, color="c", edgecolor="black")
    axes[0, 1].set_title("Histogram of Sepal Length")

    # Box Plot
    sns.boxplot(ax=axes[1, 0], data=df, x="species", y="sepal_width")
    axes[1, 0].set_title("Box Plot")

    # Line Plot
    sns.lineplot(ax=axes[1, 1], data=df, x=df.index, y="petal_width", color="r")
    axes[1, 1].set_title("Line Plot")

    plt.tight_layout()  # Adjust layout for clarity
    plt.show()

# Function call
sub_plots()

```

```python

from mpl_toolkits.mplot3d import Axes3D

# Generate Data
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))  # Function for 3D Surface

# Wireframe Plot Function
def wireframe_plot():
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection="3d")

    ax.plot_wireframe(X, Y, Z, color="b", linewidth=0.5)

    ax.set_xlabel("X Axis")
    ax.set_ylabel("Y Axis")
    ax.set_zlabel("Z Axis")
    plt.title("3D Wireframe Plot")

    plt.show()

# Function call
wireframe_plot()
```

```python

# Load dataset
df = sns.load_dataset("iris")

# Bubble Plot Function
def bubble_plot():
    plt.figure(figsize=(8, 5))
    
    # Bubble Plot (scatter plot with size variation)
    plt.scatter(df["sepal_length"], df["petal_length"], s=df["sepal_width"] * 50, alpha=0.5, c=df["petal_width"], cmap="coolwarm", edgecolors="black")
    
    plt.xlabel("Sepal Length")
    plt.ylabel("Petal Length")
    plt.title("Bubble Plot: Sepal vs Petal Length with Bubble Size as Sepal Width")
    plt.colorbar(label="Petal Width")  # Color legend for petal width
    
    plt.show()

# Function call
bubble_plot()
```

```python

# Load dataset (taking mean values for each species)
df = pd.DataFrame(sns.load_dataset("iris").groupby("species").mean())

# Features to plot
categories = df.columns.tolist()
num_vars = len(categories)

# Convert data into polar coordinates
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Repeat first angle to close the plot

# Function to plot Radar Chart
def radar_chart():
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={"projection": "polar"})

    # Plot each species
    for species in df.index:
        values = df.loc[species].tolist()
        values += values[:1]  # Repeat first value to close the plot
        ax.plot(angles, values, label=species)
        ax.fill(angles, values, alpha=0.1)  # Fill area under the line

    # Labels and title
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories)
    plt.title("Radar Chart: Iris Species Comparison")
    plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))
    plt.show()

# Function call
radar_chart()
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Define sets
set1 = {"A", "B", "C", "D"}
set2 = {"C", "D", "E", "F"}
set3 = {"D", "E", "G", "H"}

# Function to plot Venn Diagram
def venn_diagram():
    plt.figure(figsize=(5, 5))
    
    # Create Venn Diagram
    venn3([set1, set2, set3], set_labels=("Set 1", "Set 2", "Set 3"))
    
    plt.title("Venn Diagram: Overlapping Sets")
    plt.show()

# Function call
venn_diagram()
```

```python
!pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Sample Text Data
text = """
Data Science Machine Learning Python AI Deep Learning Visualization
Statistics Data Analysis Power BI SQL Pandas Matplotlib Seaborn
EDA Feature Engineering Predictive Modeling Big Data Analytics
"""

# Function to create Word Cloud
def word_cloud():
    plt.figure(figsize=(8, 5))
    
    # Generate Word Cloud without font_path
    wc = WordCloud(width=800, height=400, background_color="black", colormap="coolwarm").generate(text)
    
    # Display Word Cloud
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")  # Remove axes
    plt.title("Word Cloud: Text Frequency Visualization")
    plt.show()

# Function call
word_cloud()

```
