```python
# ðŸ”¹ Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import os

# ðŸ”¹ Step 1: Load Data
file_path = r"C:\Users\sonu kumar\Downloads\Global_sea_level_rise.csv"  # Correct path

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print("Data loaded successfully!")
else:
    print("Error: File not found!")
    exit()

# ðŸ”¹ Step 2: Clean Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print(df.head())  # Check first few rows

# ðŸ”¹ Step 3: Scatter Plot
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['year'], y=df['mmfrom1993-2008average'], label="Observed Data", color="blue")

# ðŸ”¹ Step 4: Line of Best Fit (1880-2050)
slope1, intercept1, _, _, _ = stats.linregress(df["year"], df["mmfrom1993-2008average"])
years_extended = pd.Series(range(1880, 2051))
plt.plot(years_extended, intercept1 + slope1 * years_extended, label="Best Fit (1880-2050)", color="red")

# ðŸ”¹ Step 5: Line of Best Fit (2000-2050)
df_recent = df[df["year"] >= 2000]
slope2, intercept2, _, _, _ = stats.linregress(df_recent["year"], df_recent["mmfrom1993-2008average"])
years_recent = pd.Series(range(2000, 2051))
plt.plot(years_recent, intercept2 + slope2 * years_recent, label="Best Fit (2000-2050)", color="green")

# ðŸ”¹ Step 6: Labels & Title
plt.xlabel("Year")
plt.ylabel("Sea Level (mm)")
plt.title("Rise in Sea Level")
plt.legend()
plt.grid()

# ðŸ”¹ Step 7: Save & Show
plt.savefig("sea_level_plot.png")
plt.show()
```
